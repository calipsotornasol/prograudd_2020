```python
def comuna_max_contagiados(file):
    f = open(file,'r')
    comunas=[]
    contagiados=[]
    next(f)
    for line in f:
        L=line.split(',')     
        comunas.append(L[2])
        contagiados.append(float(L[5][:-1]))
    f.close()
    
    contagiados_max=max(contagiados)
    comunas_contag_max=[]
    
    for i in range(len(contagiados)):
        if contagiados[i] == contagiados_max:
             comunas_contag_max.append(comunas[i])
    
    return comunas_contag_max
```

```python
comuna_max_contagiados('2020-05-11-CasosConfirmados.csv')
```

```python
def contagiados_comuna_list(file):
    f = open(file,'r')
    com=[]
    cont=[]
    
    next(f)
    for line in f:
        L=line.split(',')     
        com.append(L[2])
        cont.append(float(L[5][:-1]))
    f.close()
    
    return com, cont
```
